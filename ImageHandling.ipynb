# 이미지 다루기
## 이미지 읽기

```python
import cv2
```

```python
img = cv2.imread('Images/Sample_nizi1.jpg', cv2.IMREAD_GRAYSCALE)
```

cv2.imread(fileName, flag) 으로 불러 온다.
flag 는 이미지 파일을 읽을 때의 옵션.
cv2.IMREAD_COLOR : 이미지 파일을 컬러로 읽어들임. Default
cv2.IMREAD_GRAYSCALE : 흑백으로 읽어옴
cv2.IMREAD_UNCHANGED : 알파채널까지 포함하여 읽어들임
1, 0, -1로 사용 가능.

```python
img.shape
```

이미지 사이즈 : 2755 X 2141
3개의 채널로 구성됨

## 이미지 보기

```python
cv2.imshow('IMAGE color mode', img)
cv2.waitKey(0)
cv2.destroyAllWindows() # 화면에 나타난 윈도우 종료
```

## Matplotlib 사용하기

```python
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('Images/Sample_nizi2.jpg', cv2.IMREAD_COLOR)
```

```python
plt.imshow(img)
plt.xticks([])
plt.yticks([])
plt.show()
```

CV는 BGR로 이미지를 읽어오기 때문에 RGB값으로 값을 변경할 필요가 있음.

```python
b, g, r = cv2.split(img)
img2 = cv2.merge([r,g,b])

plt.imshow(img2)
plt.xticks([])
plt.yticks([])
plt.show
```

## 보너스
### 1. CV2 함수로 RGB 만들기

```python
img3 = cv2.cvtColor(cv2.imread('Images/sample_nizi2.jpg', 1), cv2.COLOR_BGR2RGB)
```

```python
img3.shape
```

```python
plt.imshow(img3)
plt.axis("off")
plt.show()
```

```python
img4 = cv2.cvtColor(cv2.imread('Images/sample_nizi2.jpg', 0), cv2.COLOR_BGR2RGB)
img5 = cv2.cvtColor(cv2.imread('Images/sample_nizi2.jpg', -1), cv2.COLOR_BGR2RGB)
```

```python
fig = plt.figure(figsize=(30,30))

ax1 = fig.add_subplot(1,3,1)
ax1.imshow(img3)
ax1.set_title("color mode")
ax1.axis("off")

ax2 = fig.add_subplot(1,3,2)
ax2.imshow(img4)
ax2.set_title("Grayscale mode")
ax2.axis("off")

ax3 = fig.add_subplot(1,3,3)
ax3.imshow(img5)
ax3.set_title("Unchanged")
ax3.axis("off")

plt.show()
```
