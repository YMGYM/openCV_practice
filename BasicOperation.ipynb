## 이미지 ROI

```python
img = cv2.cvtColor(cv2.imread('Images/Sample_nizi2.jpg'), cv2.COLOR_BGR2RGB)
```

```python
youu = img[700:1400, 100:1000]
youu.shape
```

```python
cp_img = np.array(img)
cp_img[700:1400, 1600:2500] = youu
```

```python
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(50,50))

ax1 = fig.add_subplot(1,2,1)
ax1.imshow(img)
ax1.set_title("Original")
ax1.axis("off")

ax2 = fig.add_subplot(1,2,2)
ax2.imshow(cp_img)
ax2.set_title("Duplicated")
ax2.axis("off")

plt.show()
```
